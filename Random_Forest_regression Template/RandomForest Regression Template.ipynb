```python
## Random Forest INtuition
## Step1: Pick at random K data points from the Training set
## Step2: Build the Decision Tree associated to these K data points
## Step3: Choose the number Ntree of trees you want to build and repeat Steps 1&2
## Step4: For a new data point, make each one of your Ntree trees predict the value of Y to fot rhe data ponint
## in question, and assign the new data point the avaerage across all of the predicted Y values
```

```python
##Random Forest Regression
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv("Position_Salaries.csv")
```

```python
X = dataset.iloc[:,1:2].values
y = dataset.iloc[:,2].values
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor()
```

```python
## Predicting Random Forest egression to the datasrt
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state=0)
regressor.fit(X,y)
```

```python
y_pred = regressor.predict(np.array(6.5).reshape(-1,1))
y_pred ##step이 많기 떄문에 더 accurate하다?
```

```python
##Several stairs... How to visualize this??
## Also non continuous model
```

```python
X_grid = np.arange(min(X),max(X),0.1)
X_grid = X_grid.reshape((len(X_grid),1))
plt.scatter(X,y,color='red')
plt.plot(X_grid,regressor.predict(X_grid),color='blue')
plt.title('Truth or Bluff (Random Forest Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show() ##more steps in the stair lot more splits lot more intervals..
```

```python
## 10개의 tree중에 평균을 구한겨....
## 100tree러 하면 어떻게 될까?
## Predicting Random Forest egression to the datasrt
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 100, random_state=0)
regressor.fit(X,y)

y_pred = regressor.predict(np.array(6.5).reshape(-1,1))
print(y_pred) ##step이 많기 떄문에 더 accurate하다?
X_grid = np.arange(min(X),max(X),0.1)
X_grid = X_grid.reshape((len(X_grid),1))
plt.scatter(X,y,color='red')
plt.plot(X_grid,regressor.predict(X_grid),color='blue')
plt.title('Truth or Bluff (Random Forest Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show() 
```
